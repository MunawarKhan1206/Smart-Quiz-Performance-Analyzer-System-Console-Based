```python
difficult = [
    {"question": "National flower of Pakistan?",
     "options": ["A) Rose", "B) Jasmine", "C) Tulip", "D) Lily"],
     "answer": "B"},

    {"question": "Which scientist gave laws of motion?",
     "options": ["A) Isaac Newton", "B) Galileo", "C) Einstein", "D) Tesla"],
     "answer": "A"},

    {"question": "Who died standing with a stick?",
     "options": ["A) Hazrat Suleman A.S", "B) Hazrat Musa A.S", "C) Hazrat Haroon A.S", "D) Hazrat Essa A.S"],
     "answer": "A"},

    {"question": "12 * 12?",
     "options": ["A) 144", "B) 145", "C) 331", "D) 221"],
     "answer": "A"},

    {"question": "15^2?",
     "options": ["A) 225", "B) 220", "C) 124", "D) 248"],
     "answer": "A"}
]

medium = [
    {"question": "Largest airport of Pakistan?",
     "options": ["A) Lahore", "B) Islamabad", "C) Jinnah International", "D) Peshawar"],
     "answer": "C"},

    {"question": "Largest port of Pakistan?",
     "options": ["A) Gwadar", "B) Karachi", "C) Port Qasim", "D) Pasni"],
     "answer": "B"},

    {"question": "Capital of Germany?",
     "options": ["A) Berlin", "B) Madrid", "C) Paris", "D) Rome"],
     "answer": "A"},

    {"question": "12 + 12?",
     "options": ["A) 24", "B) 21", "C) 32", "D) 15"],
     "answer": "A"},

    {"question": "Programming language?",
     "options": ["A) HTML", "B) CSS", "C) Python", "D) XML"],
     "answer": "C"}
]

easy = [
    {"question": "PC stands for?",
     "options": ["A) Private Computer", "B) Personal Computer", "C) Personal Chip", "D) Private CPU"],
     "answer": "B"},

    {"question": "Founder of Pakistan?",
     "options": ["A) Allama Iqbal", "B) Liaquat Ali Khan", "C) Quaid-e-Azam", "D) Sir Syed"],
     "answer": "C"},

    {"question": "Months in a year?",
     "options": ["A) 15", "B) 11", "C) 12", "D) 8"],
     "answer": "C"},

    {"question": "National animal of Pakistan?",
     "options": ["A) Lion", "B) Markhor", "C) Tiger", "D) Goat"],
     "answer": "B"},

    {"question": "Days in a leap year?",
     "options": ["A) 365", "B) 366", "C) 364", "D) 360"],
     "answer": "B"}
]

leaderboard = []
last_attempt = None

def performance_feedback(p):
    if p < 50:
        return "Beginner", "You need more practice."
    elif p < 75:
        return "Intermediate", "Good job!"
    else:
        return "Advanced", "Excellent!"

print("WELCOME TO SMART QUIZ & PERFORMANCE ANALYZER SYSTEM")

while True:
    print("\n1. Start Quiz\n2. Leaderboard\n3. Last Score\n4. Exit")

    
    choice = int(input("Enter option: "))
    if choice == 1:
        name = input("Enter name: ")

        print("\n1. Difficult \n2. Medium\n3. Easy")
        difficulty = int(input("Choose level: "))

        if difficulty == 1:
            questions = difficult
        elif difficulty == 2:
            questions = medium
        else:
            questions = easy

        score = 0
        correct = 0
        wrong = 0

        for q in questions:
            print("\n" + q["question"])
            for opt in q["options"]:
                print(opt)

            while True:
                ans = input("Enter answer (A/B/C/D): ").lower()
                if ans in ["a", "b", "c", "d"]:
                    break
                else:
                    print("Invalid input!")

            if ans == q["answer"].lower():
                print(f"Correct! You chose '{ans.upper()}' and that is RIGHT.")
                score += 1
                correct += 1
            else:
                print(f"Wrong! You chose '{ans.upper()}', correct answer is '{q['answer']}'")
                score -= 1
                wrong += 1

        total = len(questions)
        percentage = round((score / total) * 100, 2)

        level, feedback = performance_feedback(percentage)

        # -------- RESULT -------- #
        print("\n===== RESULT =====")
        print(f"Name: {name}")
        print(f"Correct Answers: {correct}")
        print(f"Wrong Answers: {wrong}")
        print(f"Total Questions: {total}")
        print(f"Score: {score}")
        print(f"Percentage: {percentage}%")
        print(f"Level: {level}")
        print(f"Feedback: {feedback}")

        leaderboard.append((name, score, percentage, level))
        last_attempt = (name, score, percentage, level)

    elif choice == 2:
        if not leaderboard:
            print("No attempts yet.")
        else:
            print("\n===== LEADERBOARD =====")
            i = 1
            for data in leaderboard:
                print(i, "| Name:", data[0], "| Score:", data[1], "| %:", data[2], "| Level:", data[3])
                i += 1
    elif choice == 3:
        if last_attempt:
            print("\n===== LAST SCORE =====")
            print("Name:", last_attempt[0])
            print("Score:", last_attempt[1])
            print("Percentage:", last_attempt[2])
            print("Level:", last_attempt[3])
        else:
            print("No attempts yet.")
    elif choice == 4:
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```
